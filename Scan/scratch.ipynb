```python
def findAMirror(e):
    
    EnergyToMirror = {(5,9):'Si',(9,15):'Cr',(15,20):'Rh',(20,25):'Pt'}
    
    if not 5.5 <= e <= 25:
        raise ValueError (" Energy value out of range")
        
    else:
        for erange in EnergyToMirror.keys():
            if erange[0] <= e <= erange[1]:
                return EnergyToMirror[erange]
            else:
                pass
```

```python
findAMirror(6)
```
